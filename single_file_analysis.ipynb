```python
import json
import pandas as pd
import anndata as ad
import time
import gzip
import os

print("--- Starting Spatial Data Pipeline ---")

# DEFINE FILE NAMES 
# Using sample GSM8031361
data_folder = "GSE253975_data"
matrix_file = os.path.join(data_folder, 'GSM8031361_T0081.txt.gz')
json_file = os.path.join(data_folder, 'GSM8031361_V11U14-042-A1.json.gz')

print(f"Matrix file path: {matrix_file}")
print(f"JSON file path:   {json_file}")

# LOAD SPATIAL COORDINATES (JSON) 
print(f"\nLoading spatial data from {json_file}...")
start_time = time.time()

try:
    with gzip.open(json_file, 'rt', encoding='utf-8') as f:
        spatial_data_full = json.load(f)
except FileNotFoundError:
    print(f"ERROR: File not found at {json_file}")
    print("Please check the file name and that it's in the 'GSE253975_data' folder.")
    raise

# Find the main list of all spots (longest list in the JSON)
all_spots_list = []
for key, value in spatial_data_full.items():
    if isinstance(value, list) and len(value) > 4000: # Logic to find the main spot list
        all_spots_list = value
        print(f"Found {len(all_spots_list)} total spots in JSON key: '{key}'")
        break

if not all_spots_list:
    raise ValueError("Could not find the main spot list in the JSON file.")

# FILTER BY "TRUE"
# Filter the list to get only the spots marked "tissue: true"
spatial_coords = []
for spot in all_spots_list:
    if spot.get("tissue"):  # This is the filtering step
        spatial_coords.append({
            'x': spot['imageX'],
            'y': spot['imageY']
        })

n_spots_json = len(spatial_coords)
print(f"Found {n_spots_json} in-tissue spots.")
print(f"JSON loading & filtering finished in {time.time() - start_time:.2f} seconds.\n")

# LOAD EXPRESSION MATRIX (TXT)
print(f"Loading expression matrix from {matrix_file}...")
start_time = time.time()

try:
    matrix_df = pd.read_csv(
        matrix_file,
        sep=r'\s+',        # Use whitespace as separator
        index_col=0,      # First column (genes) is the index
        engine='python',
        compression='gzip'  # Decompress .gz file
    )
except FileNotFoundError:
    print(f"ERROR: File not found at {matrix_file}")
    print("Please check the file name and that it's in the 'GSE253975_data' folder.")
    raise
except Exception as e:
    print(f"\nError loading matrix: {e}")
    raise

# Clean up the quotes from gene names and barcodes
matrix_df.columns = matrix_df.columns.str.strip('"')
matrix_df.index = matrix_df.index.str.strip('"')

n_spots_matrix = matrix_df.shape[1]
n_genes_matrix = matrix_df.shape[0]
print(f"Matrix loaded. Shape: ({n_genes_matrix} genes, {n_spots_matrix} spots)")
print(f"Matrix loading finished in {time.time() - start_time:.2f} seconds.\n")

# MATCH (Sanity Check)
# link-by-order
if n_spots_matrix != n_spots_json:
    print(f"CRITICAL WARNING: Spot counts do not match!")
    print(f"  Matrix spots: {n_spots_matrix}")
    print(f"  JSON spots:   {n_spots_json}")
    print("The pipeline will likely fail or produce an incorrect object.")
else:
    print(f"Spot counts match: (Matrix: {n_spots_matrix}, JSON: {n_spots_json})")
    print("Proceeding to create AnnData object...")

# CREATE ANNDATA OBJECT 
# AnnData format requires (observations x variables), or (spots x genes)
# Our matrix_df is (genes x spots), so we must Transpose it using .T

# Create the OBS (observations) DataFrame
# The index is the spot barcodes from the matrix columns
spot_barcodes = matrix_df.columns
obs_df = pd.DataFrame(index=spot_barcodes)

# Add coordinates to 'obs'
# We create a DataFrame from our filtered list and assign its values.
# This works *because* the order is matched.
coord_df = pd.DataFrame(spatial_coords)
obs_df['x_coord'] = coord_df['x'].values
obs_df['y_coord'] = coord_df['y'].values

# Create the VAR (variables) DataFrame
# The index is the gene names from the matrix index
gene_names = matrix_df.index
var_df = pd.DataFrame(index=gene_names)

# Create the final AnnData object
adata = ad.AnnData(X=matrix_df.T.values,  # .T transposes from (genes x spots) to (spots x genes)
                   obs=obs_df,
                   var=var_df)

# Add coordinates to the special 'obsm' slot for scanpy
# This is what plotting functions like `sc.pl.spatial()` use
adata.obsm['spatial'] = obs_df[['x_coord', 'y_coord']].values

print("\n--- AnnData Object Created Successfully! ---")
print("You can now use this object (named 'adata') for analysis.")
print(adata)
```
